```
import matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
matplotlib.__version__
#import seaborn as sns
#sns.set()
```

```
fig = plt.figure()
fig.show()
```

```
_ = plt.plot([0,1],[0,3])
```

```
plt.plot([0,1],[0,3], label='line', color="green")
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.title('Wykres')
_ = plt.legend()
```

```
plt.plot([0,1,2],[0,3,2], label='line', color="green")
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.title('Wykres')
_ = plt.legend()
```

```
x = np.arange(-3,3,0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)

plt.plot(x,x_sin)
_ = plt.plot(x,x_cos)
```

Wiele wykresów

```
plt.subplot(121) # liczba wierszy, kolumny, konkretny wykres
plt.plot([0,1,2],[3,2,5])
plt.subplot(122)
plt.plot([0,1,2],[3,1,4], 'r--')
```

```
fig = plt.figure(figsize=(10,5))

plt.subplot(131) # liczba wierszy, kolumny, konkretny wykres
plt.plot([0,1,2],[3,2,5])
plt.subplot(132)
plt.plot([0,1,2],[3,1,4], 'r--')
plt.subplot(133)
_ = plt.plot([0,1,0],[3,0,4], 'ro')
```
